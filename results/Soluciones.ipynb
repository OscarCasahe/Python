```python
"""1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias
de cada letra en la cadena. Los espacios no deben ser considerados.
"""
def frecuenciasStr(cadena:str):
    """función que recibe una cadena de texto como parámetro y devuelve un diccionario con las frecuencias
    de cada letra en la cadena

    Arg: 
        cadena(str): string a tratar para obtener las frecuencias
    Returns:
        dict: Diccionario con el resultado de frecuencias por caracter
    """
    cadena = cadena.strip().replace(" ", "")
    frecuencias = {}
        
    for caracter in cadena:
        if caracter not in frecuencias:
            frecuencias.update({caracter:1})
        else:
            frecuencias.update({caracter: frecuencias.get(caracter) + 1})
            
    return frecuencias

print(frecuenciasStr(' aaa oo iiii '))
```

```python
"""5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por
defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual
que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver
una tupla que contenga la media y el estado."""




def calificar(examenes, nota_aprobado=5):
    """ Calcula la media de los números en la lista y determinar si la media es mayor o igual
    que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso"
    Params:
        - examenes list(int o float): Lista de notas.
        - nota_aprobado (int o float): límite aprobado.
    Returns:
        tuple: tupla que contiene la media y el estado.
    """
    total = 0
    
    for nota in examenes:
        total += nota
        
    media = total / len(examenes)
    
    if(media > nota_aprobado):
        calificacion = 'aprobado'
    else:
        calificacion = 'suspenso'
    return (media, calificacion)

print(calificar([7,9,8,7]))
print(calificar([7,9,8,7], 8))
print(calificar([5,6,4,3]))
```

```python
"28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada."

def hasDup(lista: list):
    """función que busca y devuelve el primer elemento duplicado en una lista dada

    Args:
        lista (list): lista de elementos

    Returns:
        _type_: cadena de texto con el resultado de la búsqueda
    """
    lista = tuple(lista)
    for obj in lista:
        checkDup = list(lista)
        checkDup.remove(obj)
        for objDup in checkDup:
            if obj == objDup:
                return f"El elemento duplicado es: {obj}."
    
    return "No hay elementos duplicados."

print(hasDup([1, 2, "a", 2, "a"]))
print(hasDup([1, 2, "a"]))
```

```python
"""30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras
pero en diferente orden."""


def frecuenciasStr(cadena:str): # Reutilizo la función del ejercicio 1
    """función que recibe una cadena de texto como parámetro y devuelve un diccionario con las frecuencias
    de cada letra en la cadena

    Arg: 
        cadena(str): string a tratar para obtener las frecuencias
    Returns:
        dict: Diccionario con el resultado de frecuencias por caracter
    """
    cadena = cadena.strip().replace(" ", "")
    frecuencias = {}
        
    for caracter in cadena:
        if caracter not in frecuencias:
            frecuencias.update({caracter:1})
        else:
            frecuencias.update({caracter: frecuencias.get(caracter) + 1})
            
    return frecuencias

def sonAnagramas(str1: str, str2: str):
    """En esta función vamos a recurrir a la función frecuenciasStr() para obtener la cantidad de 
    repeticiones por caracter, y poder así comprobar si son anagramas

    Args:
        str1 (str): _description_
        str2 (str): _description_

    Returns:
        _type_: _description_
    """
    theyAre = True
    frecStr1 = frecuenciasStr(str1.upper()) # Obtenemos las frecuencias de ambos strings y las pasamos a mayúsculas para controlar distintos casos
    frecStr2 = frecuenciasStr(str2.upper())
    
    for charFrec1 in frecStr1.items():
        for charFrec2 in frecStr2.items():
            if charFrec1[0] == charFrec2[0] and charFrec1[1] != charFrec2[1]: # Si el caracter de ambas palabras tiene distinta frecuencia, no son anagramas
                    theyAre = False
    if theyAre:
        return "Estas cadenas son anagramas"
    else:
        return "Estas cadenas no son anagramas"

print(sonAnagramas("cosas", "sacos"))
print(sonAnagramas("amarillo", "azul"))
```

```python
"""34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son:
crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol . El objetivo es implementar estos métodos para
manipular la estructura del árbol.
Código a seguir:
1. Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
2. Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
3. Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
4. Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
5. Implementar el método quitar_rama para eliminar una rama en una posición específica.
6. Implementar el método
info_arbol para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las
mismas.
Caso de uso:
1. Crear un árbol.
2. Hacer crecer el tronco del árbol una unidad.
3. Añadir una nueva rama al árbol.
4. Hacer crecer todas las ramas del árbol una unidad.
5. Añadir dos nuevas ramas al árbol.
6. Retirar la rama situada en la posición 2.
7. Obtener información sobre el árbol."""


class Arbol:
    """Clase de Arbol
    Args:
    tronco (int): Longitud del tronco del árbol.
    puesto (str): Puesto del empleado.

    """
    
    def __init__(self, tronco, ramas):
        self.tronco = tronco
        self.ramas = list(ramas)
        
    def crecer_tronco(self):
        """Función para hacer crecer el tronco una unidad
        """
        self.tronco += 1

    def nueva_rama(self):
        """Función para incorporar una rama nueva al árbol
        """
        self.ramas.append(1)

    def crecer_ramas(self):
        """Función para hacer crecer las ramas una unidad
        """
        creceRamas = lambda ramas: [rama + 1 for rama in ramas]
        self.ramas = creceRamas(self.ramas)

    def quitar_rama(self, index: int):
        """Función para eliminar la rama del índice especificado
        """
        self.ramas.remove(self.ramas[index])

    def info_arbol(self):
        """Función obtener la información del árbol a modo d ediccionario
        """      
        return self.__dict__
    
    
arbol = Arbol(5, [5,7])

arbol.crecer_tronco()
arbol.nueva_rama()
arbol.crecer_ramas()
arbol.quitar_rama(1)
arbol.info_arbol()
```

```python
"""36. Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta
corriente. Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y
agregar dinero al saldo.
Código a seguir:
1. Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente mediante True y False .
2. Implementar el método retirar_dinero para retirar dinero del saldo del usuario. Lanzará un error en caso de no
poder hacerse.
3. Implementar el método transferir_dinero para realizar una transferencia desde otro usuario al usuario actual.
Lanzará un error en caso de no poder hacerse.
4. Implementar el método agregar_dinero para agregar dinero al saldo del usuario.
Caso de uso:
1. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
2. Agregar 20 unidades de saldo de "Bob".
3. Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
4. Retirar 50 unidades de saldo a "Alicia" """

class UsuarioBanco:
    """Clase de UsuarioBanco
    
        Args:
        nombre(str): Nombre del usuario
        saldo(float): Saldo del usuario
        cuentaCorriente(True / False): Verificador de cuenta corriente del usuario
    """
    def __init__(self, nombre, saldo, cuentaCorriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuentaCorriente = cuentaCorriente
        
    def agregarDinero(self, cantidad):
        """Función para agregar dinero

        Args:
            cantidad (float): Cantidad a incorporar

        Returns:
            str: Mensaje informativo con el saldo actual
        """
        self.saldo += cantidad
        return f"Saldo total: {self.saldo}€"
    
    def retirarDinero(self, cantidad):
        """Función para retirar dinero

        Args:
            cantidad (float): Cantidad a retirar

        Returns:
            str: Mensaje informativo con el saldo actual
        """
        if cantidad > self.saldo:
            print(f"Lo sentimos, la cantidad a retirar supera el saldo de {self.nombre}({self.saldo})")
            return False
        else:
            self.saldo -= cantidad
            return f"Saldo restante: {self.saldo}€"
        

    def transferir_dinero(self, cantidad, usuario):
        """Función para transferir dinero de un usuario a el usuario en cuestión

        Args:
            cantidad (float): Cantidad a incorporar
            usuario (UsuarioBanco): usuario externo que va a realizar la transferencia

        Returns:
            str: Mensaje informativo con el saldo actual
        """
        if usuario.retirarDinero(cantidad) != False:
            usuario.retirarDinero(cantidad)
            self.agregarDinero(cantidad)
            return f"Bizum de {cantidad}€ enviado de {usuario.nombre} a {self.nombre}"

    def infoUsuario(self):
        """Método interno de la clase para obtener la información del usuario

        Returns:
            dict: Retorna la información del usuario
        """
        return self.__dict__
    
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True) 

bob.agregarDinero(20)
alicia.transferir_dinero(80, bob)
```

```python
"""37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras ,
reemplazar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro
de la función procesar_texto .
Código a seguir:
1. Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene
que devolver un diccionario.
2. Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva. Tiene
que devolver el texto con el remplazo de palabras.
3. Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra
eliminada.
4. Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un
número de argumentos variable según la opción indicada.
Caso de uso:
Comprueba el funcionamiento completo de la función procesar_texto"""

def procesar_texto(cadena: str):
    """Función que permite seleccionar entre los métodos contar, reemplazar y eliminar para usar con una cadena de texto

    Args:
        cadena (str): cadena de texto a tratar

    Returns:
        str / dict: resultado de las funciones
    """
    response = input("¿Qué desea hacer --> contar / reemplazar / eliminar:")
    
    if response == "contar":
        return contar_palabras(cadena)
    elif response == "reemplazar":
        originalWord = input("Inserte la palabra que quiere reemplazar: ")
        replaceWord = input("Inserte la palabra por la que quiere reemplazar: ")
        
        return reemplazar_palabras(cadena, originalWord, replaceWord)
    elif response == "eliminar":
        word = input("Inserte la palabra que quiere eliminar: ")
        return eliminar_palabra(cadena, word)        

def contar_palabras(cadena: str):
    """Función para obtener la frecuencia de las palabras de una frase

    Args:
        cadena (str): cadena de texto a tratar

    Returns:
        dict: diccionario con las frecuencias de las palabras
    """
    palabras = cadena.strip().split(" ")
    frecuencias = {}
    
    for palabra in palabras:
        if palabra not in frecuencias:
            frecuencias.update({palabra:1})
        else:
            frecuencias.update({palabra: frecuencias.get(palabra) + 1})
            
    return frecuencias

def reemplazar_palabras(cadena:str, palabra_original:str, replaceWord: str):
    """Función para reemplazar dos palabras en una cadena de texto

    Args:
        cadena (str): cadena de texto a trabajar
        palabra_original (str): palabra a cambiar
        replaceWord (str): palabra cambio

    Returns:
        str: devuelve la nueva cadena de texto tras el reemplazo
    """
    return cadena.replace(palabra_original, replaceWord)

def eliminar_palabra(cadena:str, palabra:str):
    """Función para eliminar la primera instancia de una palabra en una cadena de texto

    Args:
        cadena (str): cadena de texto a tratar
        palabra (str): palabra a borrar

    Returns:
        str: devuelve la cadena de texto tras el borrado de la palabra
    """
    splitedStr = cadena.strip().split(" ") # Eliminamos los espacios de los extremos y dividimos la cadena por los espacios para crear una lista modificable
    splitedStr.remove(palabra) # Borramos la palabra en cuestión 
    result = ""
    
    for word in splitedStr: # Volvemos a crear la cadena para la respuesta
        if result != "":
            result += " " + word
        else:
            result += word
    return result


procesar_texto("Hola Buenos Días") # Aquí se pueden probar todos los casos mediante input
```
